```python
# hypothesis_test.py

import math

# Given data
sample_mean = 3050
theoretical_mean = 4000
sigma = 25
n = 25

# Calculate test statistic
standard_error = sigma / math.sqrt(n)
t_statistic = (sample_mean - theoretical_mean) / standard_error

# Determine critical value
alpha = 0.05
# standard normal z-critical value for one-tailed test
z_critical = 1.645

# Print results
print(f"Test statistic (t): {t_statistic}")
print(f"Critical value (z): {z_critical}")

# Decision
if t_statistic > z_critical:
    print("Reject the null hypothesis: costs are higher than the model suggests.")
else:
    print("Fail to reject the null hypothesis: no evidence costs are higher.")
```

```python
#Conclusion (brief):

#Based on the hypothesis test,
#there is no evidence to support the restaurant ownersâ€™ claim that the weekly operating costs are higher than the cost model predicts.
#In fact, the observed mean cost is significantly lower than the theoretical cost. 
#Therefore, we fail to reject the null hypothesis.
```

```python
import numpy as np
from scipy.stats import chi2

# observed frequencies from the table
observed = np.array([
    [50, 70],
    [80, 100],
    [60, 90],
    [30, 50],
    [20, 50]
])

# compute row totals and column totals
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

# expected frequencies
expected = np.outer(row_totals, col_totals) / grand_total

# calculate chi-square statistic
chi_square_statistic = ((observed - expected) ** 2 / expected).sum()

# degrees of freedom
df = (observed.shape[0] - 1) * (observed.shape[1] - 1)

# significance level
alpha = 0.05

# critical value from chi-square distribution
critical_value = chi2.ppf(1 - alpha, df)

# decision
if chi_square_statistic > critical_value:
    decision = "Reject the null hypothesis: There is an association."
else:
    decision = "Fail to reject the null hypothesis: No significant association."

# display results
print(f"Chi-square statistic: {chi_square_statistic:.3f}")
print(f"Critical value (alpha=0.05, df={df}): {critical_value:.3f}")
print(f"Decision: {decision}")
```

```python
#Null Hypothesis (H0):
#There is no association between the type of device purchased (Smart Thermostat or Smart Light) and customer satisfaction level.

#Alternative Hypothesis (H1): There is an association between the type of device purchased and customer satisfaction level.
```

```python
#Conclusion: There is no statistically significant association between the type of smart device purchased and the customer satisfaction level.
```
